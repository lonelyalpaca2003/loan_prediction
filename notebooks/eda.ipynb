```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../data/train.csv')
df.sample()
```

```python
df.describe()
```

```python
loan_type = df['loan_intent'].value_counts().reset_index()
loan_type
```

```python
plt.figure(figsize = (8,8))
sns.barplot(data = loan_type, x = 'count', y = 'loan_intent' , hue = 'count', palette = 'viridis')
plt.title("Counts of Loan Intent")
plt.show()
```

```python
home_own_stats = df['person_home_ownership'].value_counts().reset_index()
```

```python
plt.figure(figsize = (8,8))
sns.barplot(data = home_own_stats, x = 'person_home_ownership', y = 'count' , hue = 'person_home_ownership', palette = 'viridis')
plt.title("Counts of Loan Intent")
plt.show()
```

```python
bins = [20, 25, 30, 40, 50, 65, 125]
labels = ['20-24', '25-29', '30-39', '40-49', '50-64', '65+']
df['age_bins'] = pd.cut(df['person_age'], bins = bins, labels = labels)
```

```python
df.groupby(['age_bins'])['loan_status'].mean()
```

```python
df.info()
```

```python
df.groupby(['loan_grade'])['loan_status'].agg(average_status = 'mean', number_of_loans = 'count')
```

```python
num_cols = df.select_dtypes(["int", "float"]).columns.tolist()
corr_matrix = df[num_cols].corr()
plt.figure(figsize = (10,10))
sns.heatmap(corr_matrix, vmin = -1, vmax = 1, cmap = 'viridis', annot = True)
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.boxplot(x = df['loan_status'], y = df['person_income'], hue = df['loan_status'])
plt.show()
```
